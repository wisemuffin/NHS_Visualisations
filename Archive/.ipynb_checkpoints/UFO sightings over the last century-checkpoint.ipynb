# Analysis of UFO sightings

Data source: https://www.kaggle.com/NUFORC/ufo-sightings/data
<br><i>This dataset contains over 80,000 reports of UFO sightings over the last century.</i>

Purpose: This notebook is designed to explore the dataset prior to presenting the findings on the web.

```python
import pandas as pd
import datetime as dt
```

```python
URL = "ufo.csv" # "https://www.kaggle.com/NUFORC/ufo-sightings/downloads/scrubbed.csv"

df = pd.read_csv(URL, parse_dates = ["datetime", "date posted"])
df.head()
```

## Clean up the data types

```python
df.dtypes # check the data types in the data frame
```

clean up the dateime column which couldnt be parsed in from the read_csv() method above.

```python
# clean up the dateime column which couldnt be parsed in from the read_csv() method above.
df["datetime"] = pd.to_datetime(df["datetime"], format= '%d/%m/%Y %H:%M', errors = 'coerce')

# rename columns so we have no spaces,making it easier to work with the columns in pandas
df.rename(columns = {"duration (seconds)" : "duration_in_seconds"}, inplace=True)

#remove unessary column
del df["duration (hours/min)"]
#df = df.drop("duration (hours/min)")

# fix up the duration column
df["duration_in_seconds"] = df["duration_in_seconds"].str.replace("`", "").astype("float")

# fix up one row of data with a q in the latitude
df["latitude"] = df.latitude.str.replace("q","").astype('float')

# add a column for just the date of the sighting
df['date'] = pd.to_datetime(df['datetime'].dt.date, format= "%Y %m %d")
```

```python
# Add in the month and year of the sighting
#df['date_custom'] = df['date'].dt.floor('d') - pd.offsets.MonthBegin(1)
df['date_custom'] = df['date'].dt.floor('d') - pd.offsets.YearBegin(1)
df.head()
df_groupby_date = df.groupby(by=["date_custom"], as_index=False)
df_groupby_date.sum().head()
```

```python
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
import cufflinks as cf
import pandas as pd
import numpy as np
```

```python
plotly.offline.init_notebook_mode(connected=True) # true = offline

layout = go.Layout(
    title = "UFO Sightings",
    yaxis = dict(
        title="Sightings"
    ),
    xaxis = dict(
        title = "date"
    )
)

data = go.Data([go.Scatter(
    x = df_groupby_date.sum()["date_custom"],
    y = df_groupby_date.sum()["sightings"],
    opacity=0.5
    ) 
])



fig = go.Figure(data=data, layout=layout)

plotly.offline.iplot(fig) # shows the plot offline
```

```python
df_groupby_date = df.groupby(by=["date_custom", "country"], as_index=False)
df_groupby_date.sum().head()
df.pivot_table()
```

```python
plotly.offline.init_notebook_mode(connected=True) # true = offline

layout = go.Layout(
    title = "UFO Sightings",
    yaxis = dict(
        title="Sightings"
    ),
    xaxis = dict(
        title = "date"
    )
)

data = go.Data([go.Scatter(
    x = df_groupby_date.sum()["date_custom"],
    y = df_groupby_date.sum()["sightings"],
    opacity=0.5,
    name = cn
    )  for cn in df_groupby_date.sum("sightings")["country"]
])



fig = go.Figure(data=data, layout=layout)

plotly.offline.iplot(fig) # shows the plot offline
```
